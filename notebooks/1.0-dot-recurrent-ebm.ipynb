```python
import numpy as np
import pandas as pd
from src.algorithms import RecurrentEBM
```

```python
data = pd.read_csv('data/processed/synthetic1.csv', index_col=0)
```

```python
X = data  # .sample(2, axis=1)
X
```

```python
model = RecurrentEBM(num_epochs=15)
model.fit(X)
```

```python
results = model.predict(X)
results
```

```python
import matplotlib.pyplot as plt

plt.plot(np.arange(len(X)), X)
plt.title('Time series')
plt.show()

plt.plot(np.arange(len(X)), results)
plt.plot(len(X) * [np.mean(results) + 2*np.std(results)])
plt.title('Reconstruction error')
plt.show()
```
